```python
# Biểu đồ Histogram
from matplotlib import pyplot
from pandas import read_csv

filename = 'diabetes.csv'
data = read_csv(filename)

data.hist()
pyplot.subplots_adjust(wspace=0.3, hspace=0.5)
pyplot.show()
```

```python
# Biểu đồ độ dày (Density Plots)
from matplotlib import pyplot
from pandas import read_csv

filename = 'diabetes.csv'
data = read_csv(filename)
data.plot(kind='density', subplots=True, layout=(3,3), sharex=False)
pyplot.subplots_adjust(wspace=1, hspace=1)
pyplot.show()
```

```python
# Biểu Đồ Hộp và Dải (Box and Whisker Plots)
from matplotlib import pyplot
from pandas import read_csv

filename = "pima-indians-diabetes.data.csv"
data = read_csv(filename)
data.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False)
pyplot.show()
```

```python
# Biểu Đồ Ma Trận Tương Quan (Correlation Matrix Plot)

from matplotlib import pyplot
from pandas import read_csv
import numpy

filename = 'pima-indians-diabetes.data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(filename)

# Tính toán ma trận tương quan
correlations = data.corr()

# Vẽ biểu đồ ma trận tương quan
fig = pyplot.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)

# Thiết lập nhãn cho các trục
ticks = numpy.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)

# Hiển thị biểu đồ
pyplot.show()
```

```python
# Ma Trận Biểu Đồ Tán Xạ (Scatter Plot Matrix)

from matplotlib import pyplot
from pandas import read_csv
from pandas.plotting import scatter_matrix


filename = "pima-indians-diabetes.data.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(filename)

# Tạo ma trận biểu đồ tán xạ
scatter_matrix(data)
pyplot.show()
```
