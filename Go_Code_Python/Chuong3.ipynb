```python
# Chuỗi
data = 'hello world'
print(data[0])
print(len(data))
print(data)

# Số
value = 123.1
print(value)
value = 10
print(value)

# Boolean
a = True
b = False
print(a, b)

# Gán nhiều giá trị cùng lúc
a, b, c = 1, 2, 3

a = None
print(a)
```

```python
# Điều kiện If-Then-Else
value = 99
if value == 99:
    print('That is fast')
elif value > 200:
    print('That is too fast')
else:
    print('That is safe')
```

```python
# Vòng lặp For
for i in range(10):
    print(i)
```

```python
# Vòng lặp While
i = 0
while i < 10:
    print(i)
    i += 1
```

```python
# Tuple
a = (1, 2, 3)
print(a)

# List
mylist = [1, 2, 3]
print("Zeroth Value: %d" % mylist[0])
mylist.append(4)
print("List Length: %d" % len(mylist))
for value in mylist:
    print(value)

# Dictionary
mydict = {'a': 1, 'b': 2, 'c': 3}
print("A value: %d" % mydict['a'])
mydict['a'] = 11
print("A value: %d" % mydict['a'])
print("Keys: %s" % mydict.keys())
print("Values: %s" % mydict.values())
for key in mydict.keys():
    print(mydict[key])

# Define an array
import numpy
mylist = [1, 2, 3]
myarray = numpy.array(mylist)
print(myarray)
print(myarray.shape)


```

```python
import matplotlib.pyplot as plt
import numpy as np

myarray = np.array([1, 3, 2])
plt.plot(myarray)
plt.xlabel('x')  # Nhãn trục hoành (x)
plt.ylabel('y')  # Nhãn trục tung (y)
plt.show()
```

```python
# Biểu đồ phân tán cơ bản
import matplotlib.pyplot as plt
import numpy

x = numpy.array([1, 2, 3])
y = numpy.array([2, 4, 6])

plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Series
import numpy
import pandas

myarray = numpy.array([1, 2, 3])
rownames = ['a', 'b', 'c']
myseries = pandas.Series(myarray, index=rownames)
print(myseries)
print(myseries['a'])
```

```python
# dataframe
import numpy
import pandas
myarray = numpy.array([[1, 2, 3], [4, 5, 6]])
rownames = ['a', 'b']
colnames = ['one', 'two', 'three']
mydataframe = pandas.DataFrame(myarray, index=rownames, columns=colnames)
print(mydataframe)

print("method 1:")
print(f"one column:\n{mydataframe['one']}")
print("method 2:")
print(f"one column:\n{mydataframe.one}")
```
